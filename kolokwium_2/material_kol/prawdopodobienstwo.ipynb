```python
import numpy as np
import matplotlib.pyplot as plt

def hist_fun(probki, L_slupkow):
    rozklad_slupkow = np.arange(10, 101)
    h_slupkow = np.zeros(len(rozklad_slupkow))
    for i_probka in range(len(probki)):
        for i_slupek in range(len(rozklad_slupkow)):
            if probki[i_probka]<=rozklad_slupkow[i_slupek]:
                h_slupkow[i_slupek]+=1
                break
    return rozklad_slupkow, h_slupkow

L_slupkow = 20
N = 1000
liczby = np.arange(1000)
prubki = np.zeros(len(liczby))
for i in range(len(liczby)):
    suma_liczb = 0
    for _ in range(10):
        suma_liczb += np.random.randint(1, 10)
    prubki[i] = suma_liczb
print(prubki)
srednia = sum(liczby)/len(liczby)
warjancja = sum((x - srednia)**2 for x in liczby) / len(liczby)

rozklad_slupkow, h_slupkow = hist_fun(prubki, L_slupkow)

przekrocza_50 = 0
for i in range(len(prubki)):
    if prubki[i] > 50:
        przekrocza_50 += 1



plt.figure()
plt.bar(rozklad_slupkow, h_slupkow)
plt.xlabel("X")
plt.ylabel("P(X = xi)")
plt.title("Hisogram")
plt.show()
print(f"Srednia: {srednia}")
print(f"Warjancja: {warjancja}")
print(f"Prawdopodobieństwo, ze przekroczy 50 wynois: {przekrocza_50/N}")
```
