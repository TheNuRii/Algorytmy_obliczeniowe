```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

def cos_szereg(x, N):
    suma_szergu = 0
    for i in range(N):
        suma_szergu += (((-1) ** i)  / factorial(2 * i)) * x ** (2 * i)
    
    return suma_szergu

x = np.linspace(-10, 10, 200)
N = len(x)
y_szereg = np.zeros(N)
y = np.zeros(N)

for i in range(N):
    y_szereg[i] = cos_szereg(x[i], 10)
    y[i] = np.cos(x[i])

plt.figure()
plt.ylim([-2, 2])
plt.plot(x, y, label="cos(x) - za pomocą biblioteki numpy")
plt.plot(x, y_szereg, 'r', label="cos(x)- Taylor")
plt.xlabel('x')
plt.ylabel('f(x) = cos(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

def sin_szereg(x, N):
    suma_szergu = 0
    for i in range(N):
        suma_szergu += x**i/factorial(i)
    
    return suma_szergu

x = np.linspace(-10, 10, 200)
N = len(x)
y_szereg = np.zeros(N)
y = np.zeros(N)

for i in range(N):
    y_szereg[i] = sin_szereg(x[i], 10)
    y[i] = np.exp(x[i])

plt.figure()
plt.ylim([-2, 2])
plt.plot(x, y, label="exp(x) - za pomocą biblioteki numpy")
plt.plot(x, y_szereg, 'r', label="exp(x)- Taylor")
plt.xlabel('x')
plt.ylabel('f(x) = exp(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

#obliczyc na kartce an a bardziej roznice k
def cos_rek(x, N):
    an = 1
    y=1
    for i in range(1, N):
        an = an*((-x*x)/((2*i +2)*(2*i + 1)))
        #(-x * x/((2*i+ 2)*(2*i + 1)))
        y += an
    return y, abs(y - np.cos(x))
    

x = np.linspace(-10, 10, 300)

y_rek = np.zeros(len(x))
y_wbudowane = np.cos(x)
liczba_iteracji = np.zeros(len(x))

N = 100

for i in range(len(x)):
    y_rek[i], liczba_iteracji[i] = cos_rek(x[i], N)
    print(f"cos rekurencyjnie: {y_rek[i]}, cos: {y_wbudowane[i]}, blad: {liczba_iteracji[i]}")

fig, (ax1, ax2) = plt.subplots(2, 1)
plt.suptitle("Porównanie wykresów cos(x)")

ax1.set_ylabel("f(x) = cos(x)")
ax1.set_xlabel("x")
ax1.plot(x, y_wbudowane, label="Wykres wbudowanego cos(x)")
ax1.plot(x, y_rek, 'r',label="Wykres cos(x) - rekurencja Tay")
ax1.set_ylim([-3, 3])
ax1.legend()
ax1.grid()

ax2.plot(x, liczba_iteracji, label="wykres liczby wymaganych iteracji Taylora")
ax2.set_ylabel("liczba iteracji")
ax2.set_xlabel("x")
ax2.legend()
ax2.grid()

plt.show()
```
